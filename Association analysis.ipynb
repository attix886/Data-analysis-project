```python
import pandas as pd

log = pd.read_csv("/Users/lizihao/Desktop/数据分析项目/mysql实现/UserBehavior_200w.csv", ',', names=['user_id','item_id','category','type','timestamp'])
log.shape
```

```python
data = log[['user_id','item_id']]
data
```

```python
data = log[log.type == 'buy']
data
```

```python
data = data[['user_id','item_id']]
data
```

```python
data['item_id'] = data['item_id'].astype('str')
data_list = data.groupby('user_id').apply(lambda x:'-'.join(x['item_id']).split('-')).tolist()
data_list
```

```python
from mlxtend.preprocessing import TransactionEncoder
tr = TransactionEncoder()
tr_ary = tr.fit_transform(data_list)
```

```python
DF = pd.DataFrame(tr_ary,columns=tr.columns_)
DF
```

```python
#输出频繁项集
from mlxtend.frequent_patterns import fpgrowth

frequent_items = fpgrowth(DF, min_support=0.0001, use_colnames=True)
frequent_items
```

```python
#找到频繁项集为2的集合
frequent_items[frequent_items['itemsets'].str.len()>=2]
```

```python
frequent_items
```

```python
from mlxtend.frequent_patterns import association_rules
arresult = association_rules(frequent_items, metric='confidence', min_threshold=0.5)
```
